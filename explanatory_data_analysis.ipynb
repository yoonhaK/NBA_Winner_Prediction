# Correlation between remaining features

Features having high correlation with win shares is good, but features being strongly correlated with other features might not be helpful. Multicollinearity generally occurs when there are high correlation between two or more predictor variables.

```python
def feature_corr(x, y, **kwargs):
    cor = np.corrcoef(x, y)[0][1]
    ax = plt.gca()
    ax.annotate("r = {:.2f}".format(cor),
                xy=(.5, .1),
                xycoords=ax.transAxes,
                size = 23)


plots = sns.PairGrid(df)
plots.map(plt.scatter)
plots.map(feature_corr);
```

The scatterplots of most features are plotted randomly.

Issues:
- VORP&BPM (R = 0.89):  they are having very strong positive relationship. Box Plus Minus (BPM) is a player's contribution per 100 possesions over the league average when the player was on the court. Value Over Replacement Player (VORP) takes BPM and translates it into minutes based contribution to a team. In order to calculate VORP, you need to use BPM. Therefore, they are highly correlated. I will only use VORP.
- PER: Player Efficiency Rating is that turns a player's box score stats into a single number, so it has high correlation with other predictors. Let's exclude this one.
- G&MP: The number of playes Games (G) and Minutes Played (MP) are highly correlated with obvious reason. I will use MP.

```python
df.columns
```

```python
del df['BPM']
del df['OBPM']
del df['DBPM']
```

```python
del df['PER']
del df['G']
```

Let's look at the pair plots and correlation matrix again

```python
def feature_corr(x, y, **kwargs):
    cor = np.corrcoef(x, y)[0][1]
    ax = plt.gca()
    ax.annotate("r = {:.2f}".format(cor),
                xy=(.5, .1),
                xycoords=ax.transAxes,
                size = 23)


plots = sns.PairGrid(df)
plots.map(plt.scatter)
plots.map(feature_corr);
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 

cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(df.corr(),vmax=.7, center=0, cmap=cmap,
            square=True, linewidths=.8, cbar_kws={"shrink": .8}, ax=ax)
```

We can see the predictor variables are a lot less correlated now.
