# Data Cleaning

```python
import pandas as pd
df = pd.read_csv('player_advanced_stats.csv')
```

Let's look at the data

```python
df.head()
```

```python
df.shape
```

We can see that the table header is presented multiple times

```python
df[df.Age == 'Age'].head()
```

```python
len(df[df.Age == 'Age'])
```

## Remove multiple headers

```python
df = df.drop(df[df.Age == 'Age'].index)
```

```python
df.shape
```

## Remove Unamed Columns

We can see there are unnaamed columns with NaNs. These unnamed columns are actually the empty columns that doesn't contain any data

```python
del df['Unnamed: 19']
del df['Unnamed: 24']
```

```python
df.head()
```

```python
df.tail()
```

```python
df.to_csv('player_stats_cleaned.csv', index = False)
```
