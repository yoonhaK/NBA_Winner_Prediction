# Using pandas read_html to Web Scrape Data


```
# list of years and url link that I want to scrape
years = [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,
        2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_advanced.html'

urls = []
# Combining the URL + year strings together
for year in years:
  urls.append(url_link.format(year))
  #print(urls)

#print(urls)
len(urls)
```

## Read HTML webpage into pandas

```
# import pandas library
import pandas as pd
```

```
# create an empty list
df_list = []
# combine df with the list of data frames 
# read_html read html into list of data frames
for i in range(len(urls)):
  df_list.append(pd.read_html(urls[i],header = 0))
```

```
#to see how many list of data frames we have
len(df_list)
```

```
# to see how many data frames in each list
for i in range(len(urls)):
  print(len(df_list[i]))
```

# Combine the Data Frames

First we change the list of list of data frames to a list of data frames.

```
df = []
for i in range(len(urls)):
  df.append(df_list[i][0])
```

```
len(df)
```

## Merge DataFrames
Use pd.concat to merge a list of DataFrame into a single big DataFrame.

```
df = pd.concat(df)
df
```
