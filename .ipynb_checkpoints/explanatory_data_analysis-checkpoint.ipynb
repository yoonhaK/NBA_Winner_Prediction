### Summary Statistics

This shows the summary statistics of continuous variables in data

```python
df.describe()
```

```python
df.hist(figsize = (25,25))
plt.show()
```

Our main interest is WS (Win Shares)

```python
df.WS.hist()
```

We can see the Win Shares are centere around 0 and 1. The distribution is skewed to the right.
The odd takeaway from the distribution is the number of players with 0 win shares.

```python
df[df['G'] < 5]
```

We can see there are high number of players that barely played at all. Although these players are on NBA rosters, that doesn’t mean they will not have significant impact to both the team and their own individual stats. 

Let's update the data to only include players that had played at least 30 games and averaged at least 25 minutes of playing time per game.

```python
df = df[df['G'] > 30]

df = df[df['MP'] > 25]
```

```python
df.WS.hist()
```

There are a lot less 0 win shares players now. The distribution is still right skewed

```python
df.boxplot(column = ['WS', 'Age'])
```

We can see there are many outliers in both win shares and age
